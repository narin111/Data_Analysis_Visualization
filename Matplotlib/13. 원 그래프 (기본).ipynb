# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
values = [30, 25, 20, 13, 10, 2]
# values = [1, 1, 1, 1, 1, 1] # 비율대로 나눠짐 16.7..
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

# 원그래프 %% 두 개 붙이면 label에 % 붙는다.
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False) 
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
explode = [0.1, 0.1, 0, 0, 0, 0] # 원그래프에서 조각내기

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
explode=[0.05]*6 # 0.05가 6개 있는 리스트

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.title('언어별 선호도')
plt.legend(loc=(1.2, 0.3), title='언어별 선호도')
plt.show()
```
