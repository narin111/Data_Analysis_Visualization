# 04. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결
```

```python
x = [1, 2, 3]
y= [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, linewidth=5)
```

## Marker

```python
plt.plot(x, y, marker='o') # 데이터 있는 곳 점으로 표시 o는 영어 소문자 o
```

```python
plt.plot(x, y, marker='o', linestyle='None') 
```

```python
plt.plot(x, y, marker='v')
```

```python
plt.plot(x, y, marker='v', markersize=10) # 데이터 점 크기조정
```

```python
plt.plot(x, y, marker='X', markersize=10) 
```

- matplotlib.marker라고 검색하면 여러 종류 marker 볼 수 있음
- https://matplotlib.org/stable/api/markers_api.html

```python
plt.plot(x, y, marker='o', markersize=10, markeredgecolor='red')  # marker 테두리 색상
```

```python
plt.plot(x, y, marker='o', markersize=10, markeredgecolor='red', markerfacecolor='yellow') # marker 색상
```

## 선 스타일

```python
plt.plot(x, y, linestyle=':') # 콜론:
```

```python
plt.plot(x, y, linestyle='--') 
```

```python
plt.plot(x, y, linestyle='-.') 
```

## 색깔

```python
plt.plot(x, y, color='pink')
```

```python
plt.plot(x, y, color='#ff0000')
```

```python
plt.plot(x, y, color='b') # blue
```

```python
plt.plot(x, y, color='g') # green
```

- https://matplotlib.org/stable/gallery/color/named_colors.html

##  포맷

```python
plt.plot(x, y, 'ro--') # color:r marker:0 linestyle:--
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go') # linestyle 없음
```

```python
plt.plot(x, y, color='g', marker='o', linestyle='None')  # linestyle 없음, 위와 동일
```

## 축약어
- markderfacecolor-mfc, markersize-ms, markeredgecolor-mec, linestyle-ls
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html (표참고)

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls=":") 
```

## 투명도

```python
plt.plot(x, y, marker='o', mfc='red',ms=10, alpha=0.3) # alpha : 투명도(0~1)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5)) # 가로 세로
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi=200) # dots per inch, 확대
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x, y)
```
