# 8. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키

plt.bar(labels, values) # 막대 그래프
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키

plt.bar(labels, values, color='r') # 색상
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.5) # 각각 색상 지정, aplha-투명도
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors)
plt.ylim(175, 195) # 데이터 범위 지정 - y축 데이터 지정
```

```python
plt.bar(labels, values, color=colors, width=0.5) # 막대 두께 조절
plt.ylim(175, 195)
```

```python
plt.bar(labels, values, color=colors, width=0.5) # 막대 두께 조절
plt.ylim(175, 195)
```

```python
plt.bar(labels, values, color=colors, width=0.3) # 막대 두께 조절
plt.xticks(rotation=45) # x축의 이름 데이터 각도를 45도로 설정 - 데이터가 많아졌을 때 유용
plt.yticks(rotation=45)
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키
ticks = ['1번학생', '2번학생', '3번학생']

plt.bar(labels, values)
plt.xticks(labels, ticks)
```
