# 9. 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df=pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 185
```

```python
filt = (df['키']>=185) # 조건을 갖고 조회
df[filt]
```

```python
df[~filt] # filter의 역으로 적용 (185미만)
```

```python
df[df['키'] >= 185] # 동일한 동작
```

```python
df.loc[df['키'] >= 185, '수학'] # 키가 185 이상인 학생들의 수학 데이터
```

```python
df.loc[df['키']>=185, ['이름', '수학', '과학']]
```

## 다양한 조건

### & 그리고

```python
df.loc[df['키']>=185 & (df['학교']=='북산고') ] 
```

### | 또는

```python
df.loc[(df['키'<170])|(df['키']>200)] # 키가 170보다 작거나, 200보다 큰 학생 데이터
```

### str 함수

```python
filt = df['이름'].str.startswith('송') # '송' 씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태' 가 들어가는 사람
df[filt]
```

```python
df[~filt] # 이름에 태 들어간 사람 제외
```

```python
langs = ["Python", "Java"]
filt = df["SW특기"].isin(langs) # SW특기가 Python이거나 Java 인 사람, 대소문자 구별
df[filt]
```

```python
langs = ["python", "java"]
filt = df["SW특기"].str.lower().isin(langs) # 모두 소문자로 바꿈
df[filt]
```

```python
df["SW특기"].str.lower().isin(langs)
```

```python
df["SW특기"].str.contains('Java', na=True) # NaN 데이터에 대해서 True로 설정
```

```python
filt = df['SW특기'].str.contains('Java', na=False)
df[filt]
```
