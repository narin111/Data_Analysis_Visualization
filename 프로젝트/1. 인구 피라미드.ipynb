## 남자 데이터 정의

```python
import pandas as pd

# 시작 3개 row 무시 - 의미없는 데이터
df_m = pd.read_excel('201101_201101_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # , 없애고 int로 바꿔줌
```

```python
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w = pd.read_excel('201101_201101_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w.head(3)
```

```python
df_m.columns
```

```python
df_w.columns # 여자컬럼은 .1이 붙음
```

```python
df_w.columns = df_m.columns # 컬럼명 통일
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
```

```python
df_w
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, - df_m.iloc[0] // 1000) # 단위 : 천 명
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title('201101_201101 대한민국 인구 피라미드')
plt.savefig('201101_201101_인구피라미드.png', dpi=100)
plt.show()
```
